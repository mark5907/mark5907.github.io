### Haar transformation of the SPY price from 20110101 to 20141229 

* Remove high frequency waves for each level
* package used: <a target='_blank', href='http://www.pybytes.com/pywavelets/'>pywt</a>

    Yeah, I know. Python had everything you need to know

```python
import pywt
haar = pywt.Wavelet('haar')
coefs = pywt.wavedec(s.values, haar)

levels = 8
plt.figure(figsize=(8,4))
for i in xrange(levels):
    k = -1*(i+1)
    coefs[k] = np.zeros_like(coefs[k])
    ss = pywt.waverec(coefs, haar)
    plt.plot(ss,label='level=%d'%(i+1))

plt.legend(loc='best')
```
